```python
import pandas as pd
```

```python
price= []
name= []
comment_header=[]
comment=[]
rating=[]
date= []

for i in range(len(product6_reviews)+1):
    price.append(product6_page.div.find_all("div",{"class":"_30jeq3 _16Jk6d"})[0].text)
    name.append(product6_reviews[i].div.div.find_all("p",{"class":"_2sc7ZR _2V5EHH"})[0].text)
    comment_header.append(product6_reviews[i].div.div.div.p.text)
    comment.append(product6_reviews[i].div.find_all("div",{"class":""})[1].text)
    rating.append(product6_reviews[i].div.div.div.find_all("div",{"class":"_3LWZlK _1BLPMq"})[0].text)
    date.append(product6_reviews[i].div.find_all("p",{"class":"_2sc7ZR"})[1].text)
    
    print(price, name, comment_header, comment, rating, date )
    print()
```

```python
len(price)

price=price[0:10]   ## this i have done becouse price length was comming more, not sure why.
```

```python
## here i have converted data into dataframe to change in CSV format

review_data={
    "price":price, 
    "name": name,
    "comment_header":comment_header,
    "comment":comment,
    "rating":rating,
    "date":date
}

df=pd.DataFrame(review_data)
```
